Bar Plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[67,55,76,86]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity in thousands")
plt.title("Famous Programming Languages")
plt.bar(x,y,width=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["English","Science","Math","Social Science","Hindi"]
y=[85,92,80,95,91]
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Average marks of different subjects",fontsize=20)
plt.bar(x,y,width=0.4,color='r',edgecolor='black',linewidth=2,alpha=0.8,)
plt.legend(['Class 6th'],loc ="upper left")
plt.show()
```
