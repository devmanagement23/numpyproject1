```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[67,55,76,86]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Popularity in thousands")
plt.title("Famous Programming Languages")
plt.bar(x,y,width=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["English","Science","Math","Social Science","Hindi"]
y=[85,92,80,95,91]
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Average marks of different subjects",fontsize=20)
plt.bar(x,y,width=0.4,color='r',edgecolor='black',linewidth=2,alpha=0.8,)
plt.legend(['Class 6th'],loc ="upper left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Subjects=["python","c","c++","jaba"]
RamScore=[86,78,90,84]
KrishnaScore=[88,94,70,83]

width=0.4
RambarLocations = np.arange(len(Subjects))
KrishnabarLocations = [k+width for k in RambarLocations]

#print(ybarLocations)
#print(zbarLocations)

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Marks",fontsize=15)
plt.title("Marks Comparison",fontsize=15)

plt.bar(RambarLocations,RamScore,width,color="r")
plt.bar(KrishnabarLocations,KrishnaScore,width,color="b")

plt.xticks(RambarLocations + width/2,Subjects)

plt.legend(['RamScore','KrishnaScore'],loc ="upper left")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

Subjects=["python","c","c++","java"]
RamScore=[86,78,90,84]
KrishnaScore=[88,94,70,83]

width=0.2
RambarLocations = np.arange(len(Subjects))
KrishnabarLocations = [k+width for k in RambarLocations]

print(RambarLocations)
print(KrishnabarLocations)


plt.xlabel("Languages",fontsize=15)
plt.ylabel("Marks",fontsize=15)
plt.title("Marks Comparison",fontsize=15)

plt.bar(RambarLocations,RamScore,width,color="r")
plt.bar(KrishnabarLocations,KrishnaScore,width,color="b")

plt.xticks(RambarLocations + width/2,Subjects)

plt.legend(['RamScore','KrishnaScore'],loc ="upper left")

plt.show()
```
